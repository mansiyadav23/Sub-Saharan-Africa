```python
# Function to dynamically generate dataframe for nan trend analysis. Functionality includes,
#     1. Observe nan counts for selected countries, and custom indicators
#     2. Observe nan counts for any number of baskets with custom indicators
#     3. Choose to a math operation on chosen points of operation
#     4. Select period of observation
    
def get_df_list(df_wb, year_start, op_benchmark, countries=None, countries_indicators=None, **indicator_distributions):
    df_data = []
    row_data = []
    year_end = 2018
    
#     create list of lists for each key word argument supplied 
    if indicator_distributions.items():

        for key, value in indicator_distributions.items():
            cat = key.split("_")[0]
            _ind = key.split("_")[1::]
            ind = " ".join(_ind)
            
            for year in range(year_start, year_end, 1):
                
#                 observe nan sums over a range of chosen indices
                if cat == "basket": 
                    if len(cat) != 0:
                        nan_basket_sum = df_wb[df_wb.loc[:, "Indicator Code"].isin(value) == True].loc[:, str(year)].isna().sum()
                        df_data.append([nan_basket_sum, year, ind])

#                 apply a math operation on observed indices
                if cat == "operate":
                    if len(cat) != 0:
                        for cat_code in value:
                            df_data.append([op_benchmark(df_wb[df_wb.loc[:, "Indicator Code"] == cat_code].loc[:, str(year)]), year, bench_code])
                
#                 observe points without math operation over years, country and indice
                if cat == "observe":
                    if len(cat) != 0:
                        for cat_code in value:
                            for c_code in countries:
#                                 print(f'''{cat_code} - {year} - {c_code}''')
                                df_data.append([
                                    df_wb.loc[:, str(year)][((df_wb.loc[:, "Country Code"] == c_code) == True) &\
                                        ((df_wb.loc[:, "Indicator Code"] == cat_code) == True)].values[0],
                                    year,
                                    cat_code,
                                    code_name[c_code],
                                    code_reg[c_code]
                                ])
    
    else:
        if (len(countries)!=0) & (len(countries_indicators)!=0):
            print("Countries Triggered...")
            for year in range(year_start, year_end, 1):
                if countries:
                    for c in countries:
                        nan_country_sum = df_wb.loc[:, str(year)][((df_wb.loc[:, "Country Code"] == c) == True) &\
                                                                (df_wb.loc[:, "Indicator Code"].isin(countries_indicators) == True)].isna().sum()
#                         print(nan_country_sum)
                        df_data.append([nan_country_sum, year, code_name[c], code_reg[c]])
#                     nan_country_sum = df_wb.loc[:, str(year)][((df_wb.loc[:, "Country Code"] == c) == True) &\
#                                                                 (df_wb.loc[:, "Indicator Code"].isin(countries_indicators) == True)].isna().sum()
#                     df_data.append([nan_country_sum, year, "All Countries", "All Regions"])
        
        else:
            print("Either countries or countries_indicators are null or both not supplied")
    return df_data
```
