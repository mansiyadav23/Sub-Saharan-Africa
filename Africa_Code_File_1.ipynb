---
# Setup
---

```python
# Importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Reading the file into Python through Pandas
countries = pd.read_excel(io='wdiw.xlsx')

# Filtering [Cool Name] for our team's value: 'Jazz' and assigning it to the variable 'region'
region = countries[countries.loc[:, 'Cool Name'] == 'Jazz']
```

---
# Missing Values
---

```python
# Computing the number of null values for each column
num_nulls = region.isnull().sum()

# Computing the number of observations in our dataset
num_observations = len(region)

# Computing the percentage of null values for each column
percentage_nulls = num_nulls / num_observations

# Plotting the percentages for columns that have at least one missing value
percentage_nulls[percentage_nulls > 0].sort_values().plot(
                                                          kind='barh',
                                                          title='% of Missing Values',
                                                          figsize=(5, 15),
                                                          grid=True,
                                                          )
plt.show()
```

---
# New Sources
---

```python
# Gathered missing values from the following sources and stored it in 'WDIWEE.xlsx':
    # https://www.indexmundi.com
    # https://data.worldbank.org
    # https://data.worldbank.org/indicator/SI.DST.05TH.20

file = "WDIWEE.xlsx"

# Reading the file into Python through Pandas
Jazz = pd.read_excel(file) 
jazz_graph = pd.DataFrame.copy(Jazz)
```

---
# Analysis
---

```python
# Comparing electricity access to Urban & Rural areas 

group = 22

fig, ax = plt.subplots(figsize = (12, 5))
index = np.arange(group)
bar_width = 0.35
label_x = jazz_graph['Country Code']

rural = plt.bar(index,
             jazz_graph['Access to electricity, rural (% of rural population)'], 
             bar_width, 
             color = 'salmon',
             label = 'Rural Populations Access to Electricity',
             align = 'center')
urban = plt.bar(index+bar_width, 
             jazz_graph['Access to electricity, urban (% of urban population)'], 
             bar_width, color = 'mediumblue',
             label = 'Urban Populations Access to Electricity',
             align = 'center')

plt.xlabel('Country Code')
plt.ylabel('% of Population with Access to Electricity')
plt.title('Electricity Rural vs Urban')
plt.xticks(index+bar_width, label_x, rotation=70)
plt.yticks(np.arange(0, 125, 10))
plt.legend(loc = 'upper left')
ax.grid(False)
plt.savefig('Rural Vs Urban Populations Access to electricity.png')
plt.show()


# Comparing Age dependency ratio of young and old people

young = jazz_graph['Age dependency ratio, young (% of working-age population)']
old = jazz_graph['Age dependency ratio, old (% of working-age population)']

fig, ax = plt.subplots(figsize = (12, 5))
index = np.arange(group)
width = 0.50

g3 = plt.bar((jazz_graph['Country Code']), 
             young, 
             width, 
             color = 'red')
g4 = plt.bar((jazz_graph['Country Code']), 
             old, 
             width, 
             color = 'black', 
             bottom = young.tolist())

plt.xlabel('Country Code')
plt.ylabel('# of Dependents(0-14 & 65+) on the Working Population')
plt.title('Age Dependency Ratio')
plt.xticks(index, (jazz_graph['Country Code']), rotation=70)
plt.yticks(np.arange(0, 115, 10))
plt.legend((g3[0], g4[0]), ('Young', 'Old'), loc = 'upper left')
ax.grid(False)
plt.savefig('Non-Working age population.png')
plt.show()

# Plotting the military Expenditure

x = jazz_graph['Country Code']
y = jazz_graph['Military expenditure (% of GDP)']
fig, ax = plt.subplots(figsize = (12, 5))

plt.xlabel('Country Code')
plt.ylabel('Military Expenditure')
plt.xticks(rotation=70)
plt.title('Military Expenditure')
plt.bar(x, y, color = 'royalblue')
ax.grid(False)
plt.savefig('Military Expenditure.png')
plt.show()

# Electricity Urban - Rural Graph 

group = 22

fig, ax = plt.subplots(figsize = (12, 5))
index = np.arange(group)
bar_width = 0.35


high = plt.bar(index, 
             jazz_graph['Income share held by highest 20%'], 
             bar_width, color = 'darkorange',
             label = 'Income share held by the highest 20%',
             align = 'center')
low = plt.bar(index+bar_width, 
             jazz_graph['Income share held by lowest 20%'], 
             bar_width, color = 'blue',
             label = 'Income share held by the lowest 20%',
             align = 'center')


plt.xlabel('Country Code')
plt.ylabel('Income')
plt.title('Highest vs Lowest Income')
plt.xticks(index+bar_width, 
           (jazz_graph['Country Code']), rotation=70)
plt.yticks(np.arange(0, 70, 10))
plt.legend(loc = 'upper left')
ax.grid(False)
plt.savefig('Highest and lowest income.png')
plt.show()


# Comparing the GDP 

x = jazz_graph['Country Code']
y = jazz_graph['GDP (current US$)']
fig, ax = plt.subplots(figsize = (12, 5))

plt.xlabel('Country Code')
plt.ylabel('GDP (current US$)')
plt.xticks(rotation=70)
plt.title('GDP (current US$)')
plt.bar(x, y, color = 'red')
ax.grid(False)
plt.savefig('GDP (current US$).png')
plt.show()


# comparing the GDP annual %

x = jazz_graph['Country Code']
y = jazz_graph['GDP growth (annual %)']
fig, ax = plt.subplots(figsize = (12, 5))

plt.xlabel('Country Code')
plt.ylabel('GDP growth (annual %)')
plt.xticks(rotation=70)
plt.title('GDP growth (annual %)')
plt.bar(x, y, color = 'purple')
ax.grid(False)
plt.savefig('GDP growth (annual %).png')
plt.show()

```
